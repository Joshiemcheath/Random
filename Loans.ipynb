```python
# NOTE: MINIMUM PAYMENTS ALWAYS TAKE PRIORITY
# Subtract minimum payments from principals before inserting into loan_amounts

#The principal balances of each loan
loan_amounts = [6082.65,
                23632,
                580,
                7111.38,
                7951.45,
                5500,
                2022.08
               ]


#The annual rate of each loan
rates = [0.125,
         0.1059,
         0.1099,
         0.07,
         0.07,
         0.07,
         0.07
        ]


#How often each loan is compounded (Student loans are typically compounded daily)
compounds_per_year = [365, 365, 365, 365, 365, 365, 365] 

#The amount that you can pay
amount_to_spend = 3000

num_loans = len(loan_amounts)
should_pay = [0] * num_loans

#While we still have money to spend AND we haven't paid off all the loans...
while amount_to_spend > 0 and sum(loan_amounts) > 0:
    
    #Calculate how much interest each loan would accrue at the next compound
    growth = [loan_amounts[i] * rates[i] / compounds_per_year[i] for i in range(num_loans)]
    
    #Pay a penny towards the fastest growing loan
    loan_to_pay_index = growth.index(max(growth))
    should_pay[loan_to_pay_index] = round(should_pay[loan_to_pay_index] + 0.01, 2)
    loan_amounts[loan_to_pay_index] = round(loan_amounts[loan_to_pay_index] - 0.01, 2)
    amount_to_spend = round(amount_to_spend - 0.01, 2)

    
#Print out how much should be paid towards each loan
for i, val in enumerate(should_pay):
    print("Pay " + str(val) + " towards loan #" + str(i))
```
